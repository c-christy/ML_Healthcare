```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

plasma_glucose1314= pd.read_sas('GLU_H.XPT')
plasma_glucose1516= pd.read_sas('GLU_I.XPT')
```

```python
plasma1=plasma_glucose1314[['SEQN','LBXGLU']]
plasma2=plasma_glucose1516[['SEQN','LBXGLU']]
```

```python
plasma1=plasma1.dropna()
plasma2=plasma2.dropna()
```

```python
#plasma=plasma['LBXGLU']
```

```python
plasma1.loc[(plasma1.LBXGLU<=126),"LBXGLU"]=0
plasma2.loc[(plasma2.LBXGLU<=126),"LBXGLU"]=0
```

```python
plasma1.loc[(plasma1.LBXGLU>126),"LBXGLU"]=1
plasma2.loc[(plasma2.LBXGLU>126),"LBXGLU"]=1
```
