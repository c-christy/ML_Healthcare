```python
import pandas as pd
```

```python
mc1314 = pd.read_sas('MCQ_H.XPT')
mc1314 = mc1314.iloc[:,[0,18,21,31]]
mc1516 = pd.read_sas('MCQ_I.XPT')
mc1516 = mc1516.iloc[:,[0,13,16,26]]
```

```python
mc1314.MCQ160A = mc1314.MCQ160A.fillna(10)
mc1314.MCQ160N = mc1314.MCQ160N.fillna(10)
mc1314.MCQ160F = mc1314.MCQ160F.fillna(10)
```

```python
mc1516.MCQ160A = mc1516.MCQ160A.fillna(10)
mc1516.MCQ160N = mc1516.MCQ160N.fillna(10)
mc1516.MCQ160F = mc1516.MCQ160F.fillna(10)
```

```python
mc1314['WalkingDisability1'] = 0
mc1314.loc[(mc1314.MCQ160A == 1)|(mc1314.MCQ160N == 1)|(mc1314.MCQ160F == 1), "WalkingDisability1"] = 1
mc1314.loc[(mc1314.MCQ160A == 10)&(mc1314.MCQ160N == 10)&(mc1314.MCQ160F == 10), "WalkingDisability1"] = 10
indexNames1314 = mc1314[mc1314['WalkingDisability1'] == 10].index
mc1314.drop(indexNames1314, inplace = True)
```

```python
mc1516['WalkingDisability1'] = 0
mc1516.loc[(mc1516.MCQ160A == 1)|(mc1516.MCQ160N == 1)|(mc1516.MCQ160F == 1), "WalkingDisability1"] = 1
```

```python
mc1314.to_csv('MedicalCondition1314.csv', index=False)
mc1516.to_csv('MedicalCondition1516.csv', index=False)
```
