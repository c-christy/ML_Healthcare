```python
import pandas as pd

kidney = pd.read_csv('kidney_y.csv')
```

```python
from imblearn.over_sampling import SMOTENC
```

```python
y=kidney.loc[:,"KIQ022"]
```

```python
y=y.to_frame("KIQ022")
```

```python
X=kidney.iloc[:,2:338]
```

```python
X
```

```python
import numpy as np
arr =np.arange(300)
```

```python
arr
```

```python
smote_nc = SMOTENC(categorical_features=arr, random_state=0)
X_resampled, y_resampled = smote_nc.fit_resample(X, y)
```

```python
x=pd.DataFrame(X_resampled)
```

```python
y=pd.DataFrame(y_resampled)
```
